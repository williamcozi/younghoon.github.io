```python
review_list = []
for page in range(1,11):
    url = f'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=161967&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={page}'
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    for i in range(10):
        review = soup.find('span',{'id':f'_filtered_ment_{i}'})
        review = review.get_text().strip()
        review_list.append(review)
review_list

with open('./naver_movie_request_2000.txt','w',encoding='utf-8') as f:
    for single_review in review_list:
        f.write(single_review+'\n')

f = open("./naver_movie_request_2000.txt", "r") 
lines = f.read()

##리스트를 텍스트 파일로 변환한 다음 다시 그 텍스트 파일을 구글 colab에서 읽게 했다. 
```

```python
print(lines)
```

```python
nouns = okt.nouns(lines)
print(nouns)
```

```python
from collections import Counter
count = Counter(nouns)

temp = count.copy()
for i in temp.keys():
  if len(i) == 1:
    del count[i]

print(count)

#counter 함수를 써서 각 원소의 개수를 딕셔너리로 변환
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf', background_color='white', width=800, height=800).generate_from_frequencies(count) 

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)  
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()  
```
